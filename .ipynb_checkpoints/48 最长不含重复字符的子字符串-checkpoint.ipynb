# 方法1 动态规划 + 哈希表

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        #动态规划
        #记录s[j]结尾的 “最长不重复子字符串” 的长度
        n = len(s)
        if n <= 1:
            return n
        dp = [0 for _ in range(n+1)]
        #记录每个字符出现的最后位置
        dic = {}
        for i in range(1,n+1,1):
            if s[i-1] not in dic:
                dic[s[i-1]] = i 
                dp[i] = dp[i-1] + 1
            elif dp[i-1] < i - dic[s[i-1]]:
                dic[s[i-1]] = i 
                dp[i] = dp[i-1] + 1
            else:
                dp[i] = i - dic[s[i-1]]
                dic[s[i-1]] = i
        return max(dp)
```

时间复杂度：O(N)  
空间复杂度：O(N)

## 空间复杂度优化

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        n = len(s)
        if n <= 1:
            return n
        res, tmp = 0,0
        dic = {}
        for j in range(n):
            i = dic.get(s[j],-1)
            dic[s[j]] = j
            if tmp < j - i:
                tmp += 1
            else:
                tmp = j - i
            res = max(res,tmp)
        return res
```

# 方法2 双指针 + 哈希表

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        #双指针+哈希
        n = len(s)
        if n <= 1:
            return n
        dic = {}
        res,i=0,-1
        for j in range(n):
            if s[j] in dic:
                i = max(i,dic[s[j]])
            dic[s[j]] = j
            res = max(res, j-i)
        return res
```
